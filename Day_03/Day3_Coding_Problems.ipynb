```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

student1 = Student("Chinkki", 20)
student1.display()
```

```python
class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def bark(self):
        print("I bark")

class Lion(Animal):
    def roar(self):
        print("I roar")
```

INHERITANCE

```python
class Animal:

    def eat(self):
        print("Animal is eating.")

class Dog(Animal):

    def bark(self):
        print("Bow Bow!")

class Cat(Animal):

    def meow(self):
        print("Meow Meow!")

dog = Dog()
cat = Cat()

dog.eat()      
dog.bark()    



cat.eat()      
cat.meow()    
```
