```python
n = int(input())
print("Enter a number:",n)
if n <= 1:
    print("Not prime")
else:
    prime = True
    for i in range(2,n):
        if n % i == 0:
            prime = False
            break;
if prime:
    print(f"{n} is a Prime number")
else:
    print(f"{n} is not a Prime number")
```

```python
#List comprehension
numbers = [1,2,3,4,5]
squares = [x*x for x in numbers]
print("Square of values in list:", squares)
numbers = range(10)
even = [x for x in numbers if x%2==0]
print("Even numbers in the list:",even)
names = ["john","mary","alex"]
upper = [name.upper() for name in names]
print("Values of list in uppercase:",upper)
```

```python
#set comprehension
numbers = [1,2,3,4,5]
squares = {x*x for x in numbers}
print("Square of values in set:", squares)

numbers = range(10)
even = {x for x in numbers if x%2==0}
print("Even numbers in the set:",even)

names = ["john","mary","alex"]
upper = {name.upper() for name in names}
print("Values of set in uppercase:",upper)
```
