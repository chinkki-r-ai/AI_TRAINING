## 2. Reading the Raw Data (CSV)

```python
df = pd.read_csv("customers_raw.csv")
print("Shape:", df.shape)
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

## 3. Detecting & Handling Missing Values

```python
print("Missing values per column:\n")
print(df.isnull().sum())

print("\nRows with ANY missing value:")
df[df.isnull().any(axis=1)]
```

```python
# Strategy: numeric columns -> fill with median 
df["Age"] = df["Age"].fillna(df["Age"].median())
df["Income"] = df["Income"].fillna(df["Income"].median())
df["Purchase_Amount"] = df["Purchase_Amount"].fillna(df["Purchase_Amount"].median())

print("Missing values after filling:\n")
print(df.isnull().sum())
```

## 4. Removing Duplicates & Cleaning Text Columns

```python
# Check & drop duplicate rows
print("Duplicate rows found:", df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print("Shape after dropping duplicates:", df.shape)
```

```python
# Clean up City column: strip whitespace, standardize casing
print("Unique City values BEFORE cleanup:", df["City"].unique())

df["City"] = df["City"].str.strip().str.title()

print("Unique City values AFTER cleanup:", df["City"].unique())
```

```python
# City also had some missing (None) values - fill with the mode (most frequent city)
city_mode = df["City"].mode()[0]
df["City"] = df["City"].fillna(city_mode)
print("Missing City values remaining:", df["City"].isnull().sum())
```

## 5. Filtering Rows

```python
high_income = df[df["Income"] > 60000]
print("High income customers:", len(high_income))
high_income[["Name", "City", "Income"]]
```

```python
# Multiple conditions combined with & and |
mumbai_high_earners = df[(df["City"] == "Mumbai") & (df["Income"] > 50000)]
mumbai_high_earners[["Name", "City", "Income"]]
```

```python
# Using .isin()
metro_cities = ["Delhi", "Mumbai", "Bangalore"]
metro_customers = df[df["City"].isin(metro_cities)]
print("Customers in metro cities:", len(metro_customers))
```

## 6. GroupBy — Split, Apply, Combine

```python
city_summary = df.groupby("City").agg(
    avg_income=("Income", "mean"),
    avg_loan=("Loan_Amount", "mean"),
    customer_count=("CustomerID", "count")
).round(2)

city_summary.sort_values("avg_income", ascending=False)
```

```python
# Which city has the highest average purchase amount?
top_city = df.groupby("City")["Purchase_Amount"].mean().idxmax()
print("City with highest average purchase amount:", top_city)
```


## 7. Data Preprocessing (Encoding + Scaling)

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

le = LabelEncoder()
df["City_encoded"] = le.fit_transform(df["City"])
print(dict(zip(le.classes_, le.transform(le.classes_))))

df[["City", "City_encoded"]].drop_duplicates()
```

```python
scaler = StandardScaler()
numeric_cols = ["Age", "Income", "Loan_Amount", "Purchase_Amount"]
scaled_values = scaler.fit_transform(df[numeric_cols])

scaled_df = pd.DataFrame(scaled_values, columns=[c + "_scaled" for c in numeric_cols])
df = pd.concat([df, scaled_df], axis=1)

df[[c for c in df.columns if "scaled" in c]].head()
```

## 8. Feature Engineering

```python
df["Debt_to_Income"] = (df["Loan_Amount"] / df["Income"]).round(3)

df["Join_Date"] = pd.to_datetime(df["Join_Date"])
today = pd.Timestamp("2026-07-16")
df["Tenure_Years"] = ((today - df["Join_Date"]).dt.days / 365).round(1)

df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 25, 35, 50, 100],
    labels=["<=25", "26-35", "36-50", "50+"]
)

df[["Name", "Age", "Age_Group", "Debt_to_Income", "Tenure_Years"]].head(10)
```

## 9. Final Cleaned Dataset

```python
print("Final shape:", df.shape)
print("\nMissing values check:\n", df.isnull().sum())
df.head()
```
