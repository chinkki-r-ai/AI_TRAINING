```python
class InventoryTransaction:
    def __init__(self, store):
        self.store = store
 
    def __enter__(self):
        self.temp = self.store.copy()
        return self.temp
 
    def __exit__(self, exc_type, exc, tb):
        if exc_type is None:
            self.store.clear()
            self.store.update(self.temp)
        return True  # Suppress exceptions
 
 
def simulate_transaction(initial_stock, operations):
    store = initial_stock.copy()
 
    with InventoryTransaction(store) as stock:
        for action, item, qty in operations:
            if action == "add":
                stock[item] = stock.get(item, 0) + qty
            elif action == "remove":
                if stock.get(item, 0) < qty:
                    raise InsufficientStockError
                stock[item] -= qty
            else:
                raise ValueError
 
    return store
```
