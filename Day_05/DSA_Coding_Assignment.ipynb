EXAMPLE 1

```python
n = int(input("Enter a number: "))
for i in range(n):
    print(i)
```

If:

n = 5 → loop runs 5 times
n = 100 → loop runs 100 times
n = 1000 → loop runs 1000 times

Operations increase linearly.

Time Complexity:

O(n)

EXAMPLE 3

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

If n = 5

5 × 5 = 25 operations

If n = 100

100 × 100 = 10,000 operations

Time Complexity:

O(n²)
