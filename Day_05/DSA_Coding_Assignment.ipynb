ARRAYS

```python
#LINEAR TRAVERSAL OF AN ARRAY
arr = [1, 2, 3, 4, 5]

print("Linear Traversal: ", end=" ")
for i in arr:
    print(i, end=" ")
print()
```

```python
#REVERSE TRAVERSAL OF AN ARRAY
arr = [1, 2, 3, 4, 5]

print("Reverse Traversal: ", end="")
for i in range(len(arr) - 1, -1, -1):
    print(arr[i], end=" ")
print()
```

```python
#TRAVERSAL OF AN ARRAY USING FOR LOOP
arr = [10, 20, 30, 40, 50]

print("Traversal using for loop: ", end=' ')
for i in arr:
    print(i, end=' ')
print()
```

```python
#TRAVERSAL OF AN ARRAY USING WHILE LOOP
arr = [1, 2, 3, 4, 5]

n = len(arr)
i = 0

print("Traversal using while loop: ", end=" ")
while i < n:
    print(arr[i], end=" ")
    i += 1
print()
```

```python
#TRAVERSAL OF AN ARRAY USING FOREACH LOOP
arr = [11,22, 33, 44, 55]

print("Traversal using foreach loop:", end=' ')
for value in arr:
    print(value, end=' ')
print()
```

```python
#SEARCHING ELEMENT IN AN ARRAY
arr = [10, 20, 30, 40, 50]
target = 30
found = False

# Linear search using traversal
for i in range(len(arr)):
    if arr[i] == target:
        found = True
        break

if found:
    print("Element found!")
else:
    print("Element not found!")
```

```python
#MODIFYING ELEMENTS IN AN ARRAY
arr = [10, 20, 30, 40, 50]

# Modifying elements using traversal (increasing each by 5)
for i in range(len(arr)):
    arr[i] += 5

# Print modified array
print("Modified array:", end=' ')
for num in arr:
    print(num, end=' ')
print()
print(len(arr))
```

```python
#STACH USING PYTHON LIST
stack = []

stack.append(10)
stack.append(20)
stack.append(30)

print("Stack after Push:", stack)
removed = stack.pop()
print("Popped Element:", removed)

print("Stack after Pop:", stack)
print("Top Element:", stack[-1])

if len(stack) == 0:
    print("Stack is Empty")
else:
    print("Stack is Not Empty")
```

```python
#LINEAR SEARCH 
arr = [10, 20, 30, 40, 50]
target = 40

for i in range(len(arr)):
    if arr[i] == target:
        print("Element found at index", i)
        break
else:
    print("Element not found")
```

```python
#BINARY SEARCH
arr = [10, 20, 30, 40, 50, 60, 70]
target = 50

low = 0
high = len(arr) - 1

while low <= high:
    mid = (low + high) // 2

    if arr[mid] == target:
        print("Element found at index", mid)
        break

    elif arr[mid] < target:
        low = mid + 1

    else:
        high = mid - 1
```

```python
#BUBBLE SORT
arr = [5, 3, 8, 1]

n = len(arr)

for i in range(n):
    for j in range(n - i - 1):
        if arr[j] > arr[j + 1]:
            arr[j], arr[j + 1] = arr[j + 1], arr[j]

print(arr)
```

RECURSION

```python
def factorial(n):
    if n == 1:
        return 1          # Base Case
    return n * factorial(n - 1)

print(factorial(4))
```

```python
def fun(n):
    if n == 0:
        return 0
    return n + fun(n-1)

print(fun(3))
```
