SCALARS, VECTORS, MATRICES, TENSORS

```python
scalar = 10
print("Scalar:", scalar)

vector = np.array([1, 2, 3])
print("\nVector:")
print(vector)

matrix = np.array([[1, 2],
                   [3, 4]])
print("\nMatrix:")
print(matrix)

tensor = np.array([
    [[1,2],[3,4]],
    [[5,6],[7,8]]
])

print("\nTensor:")
print(tensor)
```

MATRIX MULTIPLICATION

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([[5,6],
              [7,8]])

print("Matrix Multiplication:")
print(np.matmul(A,B))
```

EIGENVALUES

```python
matrix = np.array([[2,0],
                   [0,3]])

values, vectors = np.linalg.eig(matrix)

print("Eigenvalues:")
print(values)

print("\nEigenvectors:")
print(vectors)
```

MEAN, MEDIAN, VARIANCE, STANDARD_DEVIATION

```python
data = np.array([10,20,30,40,50])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Variance:", np.var(data)) #find the mean, subtarct each value from the mean, square the result, and then find the mean of those squared differences
print("Standard Deviation:", np.std(data)) #square of variance, which is the square root of the variance
```

CORRELATION

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

print("Correlation Matrix:")
print(np.corrcoef(x,y))
```

NORMAL DISTRIBUTION

```python
data = np.random.normal(0,1,1000)

print("First 10 values:")
print(data[:10])

print("Mean:", np.mean(data))
print("Standard Deviation:", np.std(data))
```
