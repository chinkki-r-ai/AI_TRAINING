```python
import sqlite3

conn = sqlite3.connect('students.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    grade TEXT
)
''')
conn.commit()
print("Table created")
```

```python
cursor.execute("INSERT INTO students VALUES (1, 'Meera', 20, 'A')")
cursor.execute("INSERT INTO students VALUES (2, 'Arjun', 22, 'B')")
cursor.execute("INSERT INTO students VALUES (3, 'Zara', 19, 'A')")
conn.commit()
print("Rows inserted")
```

```python
cursor.execute("SELECT * FROM students")
print(cursor.fetchall())
```

```python
cursor.execute("UPDATE students SET grade = 'A+' WHERE name = 'Arjun'")
conn.commit()

cursor.execute("SELECT * FROM students")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM students WHERE name LIKE 'M%'")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM students WHERE age BETWEEN 20 AND 25")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT * FROM students WHERE grade IN ('A', 'A+')")
print(cursor.fetchall())
```

```python
cursor.execute("INSERT INTO students VALUES (4, 'Rohan', NULL, 'B')")
conn.commit()

cursor.execute("SELECT * FROM students WHERE age IS NULL")
print(cursor.fetchall())
```

```python
cursor.execute("SELECT DISTINCT grade FROM students")
print(cursor.fetchall())
```

```python
cursor.execute("DELETE FROM students WHERE name = 'Zara'")
conn.commit()

cursor.execute("SELECT * FROM students")
print(cursor.fetchall())
```
