### 1. MATPLOTLIB

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

plt.plot(x, y, marker='*')
plt.title("My First Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

CORRELATION HEATMAPS

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Hours_Studied': [1, 2, 3, 4, 5, 6, 7],
    'Marks_Scored': [50, 55, 65, 70, 78, 85, 95],
    'Sleep_Hours': [8, 7, 6, 6, 5, 5, 4]
}
df = pd.DataFrame(data)

correlation_matrix = df.corr()
print(correlation_matrix.round(2))

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Restaurant performance data (7 days)

data = {
    "Tables_Booked": [30, 42, 38, 50, 45, 60, 55],
    "Bill_Amount": [65, 72, 68, 80, 75, 90, 85],
    "Wait_Time": [12, 18, 15, 25, 20, 30, 28]
}

df = pd.DataFrame(data)
print("Restaurant Data:\n")
print(df)
correlation = df.corr()
print("\nCorrelation Matrix:\n")
print(correlation.round(2))
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap="coolwarm")
plt.title("Restaurant Performance Correlation Heatmap")
plt.show()
```

### VISULAIZATION USING DATASET

```python
import seaborn as sns

diamonds = sns.load_dataset('diamonds')
print(diamonds.head())
print(diamonds.columns)
print(diamonds.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

plt.hist(diamonds['price'], bins=30, color='green', edgecolor='black')
plt.title("Distribution of Diamond Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(diamonds['carat'], diamonds['price'], alpha=0.3, color='purple')
plt.title("Carat vs Price")
plt.xlabel("Carat")
plt.ylabel("Price")
plt.show()
```

```python
numeric_cols = diamonds[['carat', 'price', 'depth', 'table']]
correlation = numeric_cols.corr()
print(correlation.round(2))

sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Diamond Attributes Correlation Heatmap")
plt.show()
```
